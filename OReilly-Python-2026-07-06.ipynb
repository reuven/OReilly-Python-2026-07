# What is a programming language?

We're here to learn Python. Python is a programming language. But what does that even mean?

When computers were first invented, they could solve one problem at a time. If you wanted to change the problem that a computer was solving, you needed to rebuild the computer, or get a new one. 

Pretty soon, people realized that it would make more sense to create a general-purpose, programmable computer that could solve many different types of problems. How did we give it instructions? We would use 1s and 0s, the so-called "binary code."

Programming in binary was very difficult and tedious (and error prone!). Pretty soon, people created programming languages. The idea was that you would write in a language, and then you would use a program to translate your program into 1s and 0s. You can think of this as like getting a book translated.

The programming languages that people created had different advantages and disadvantages. Some ran really fast. Some were easy to understand. Some were great for math. Some were great for graphics. etc.

There are, today, many thousands of programming languages. Some are:

- C and C++: These run super fast, in part because they are very close to the 1s and 0s, making it hard to write with these languages, but they're super efficient.
- Java and C#: Easier to work with than C and C++, but not quite as fast. Also, automatic management of memory, which is a big deal, since in C, you have to keep track of what memory you're using.
- Python was deliberately invented to give everyone the power of a serious programming language without the very difficult learning curve. Python has proven itself over the years to be a serious language, but also as easy as it can be to learn.

- Low floors and high ceilings

# Variables and assignment

It's annoying and boring to use the same values repeatedly. What we want to do is allow for different values in our program. For that, we need *variables*. 

You can think of a variable as a pronoun, referring to a piece of data (a noun). We can assign a value to a variable with the `=` "assignment operator."

A few things to keep in mind:

1. The `=` in Python is **NOT** doing what the `=` in math does. It's very very different.
2. When you assign, Python looks at the right side, and gets the value.
3. It then assigns that value to the variable on the left side.
4. If the variable didn't previously exist, now it does!
5. If the variable *did* previously exist, now it refers to a new value.

```python
x = 10    # here, I'm assigning the value 10 to the variable x

print(x + 5)
```

```python
print(x + 20)
```

```python
print(x + x)
```

In many languages, you need to *declare* your variables. Not so in Python! The first time you assign to a variable is when it is created.

```python
print('abcd' + 'efgh' + 'ijkl')
```

```python
print(10 + 20 + 30)
```

# Things to keep in mind about variable names

1. Variable names are for *you*, not for Python. You should choose meaningful names.
2. You can use any letters, any digits, and `_` in a variable name.
3. Capital letters and lowercase letters are different.
4. Traditionally, we only use lowercase letters (not capitals) in variable names.
5. You shouldn't start or end a variable name with `_`.
6. You cannot start with a digit.



```python
x = 10       # assign the integer (number) 10 to x
x = 'abcd'   # assign the text string 'abcd' to x

print(x)     # the most recent assignment wins!
```

# Comparison operators

We've seen one operator so far, `+`, which allows us to add together two numbers (and get a new number) or add together two text strings (and get a new text string).

But there are lots of other operators in Python! One of them is `==`, the "equality operator."

If we put values on either side of `==`, we'll get back one of the two special values `True` or `False`, indicating whether they're equal.

```python
10 == 10
```

```python
10 == 5
```

```python
'abcd' == 'abcd'
```

# There are lots of comparison operators in Python

Each of these compares two values, and returns `True` or `False`, reflecting the comparison.

- `==` - the equality comparison
- `!=` -- inequality, the opposite of `==`
- `<` -- less than
- `<=` -- less than or equal to
- `>` -- greater than
- `>=` -- greater than or equal to

```python
5 < 10
```

```python
100 < 200
```

```python
'abcd' < 'efgh'
```

If we compare (with `<` or `>`) text strings, the comparison is based on their alphabetical order (assuming that it's just lowercase letters).

# Conditions

Almost all of programming, when you come down to it, is about making decisions. We can now do that!

We use `if` to make a decision. It looks to its right, and looks for either `True` or `False`. That value (`True` or `False`) usually comes from a comparison operator.

If the `if` condition is `True`, then a block of code runs. But that block of code will *only* run if it's `True`. This is a completely new idea in our programs -- so far, everything has run, from start to finish. But now, only parts of our programs will run.
