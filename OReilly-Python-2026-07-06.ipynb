# What is a programming language?

We're here to learn Python. Python is a programming language. But what does that even mean?

When computers were first invented, they could solve one problem at a time. If you wanted to change the problem that a computer was solving, you needed to rebuild the computer, or get a new one. 

Pretty soon, people realized that it would make more sense to create a general-purpose, programmable computer that could solve many different types of problems. How did we give it instructions? We would use 1s and 0s, the so-called "binary code."

Programming in binary was very difficult and tedious (and error prone!). Pretty soon, people created programming languages. The idea was that you would write in a language, and then you would use a program to translate your program into 1s and 0s. You can think of this as like getting a book translated.

The programming languages that people created had different advantages and disadvantages. Some ran really fast. Some were easy to understand. Some were great for math. Some were great for graphics. etc.

There are, today, many thousands of programming languages. Some are:

- C and C++: These run super fast, in part because they are very close to the 1s and 0s, making it hard to write with these languages, but they're super efficient.
- Java and C#: Easier to work with than C and C++, but not quite as fast. Also, automatic management of memory, which is a big deal, since in C, you have to keep track of what memory you're using.
- Python was deliberately invented to give everyone the power of a serious programming language without the very difficult learning curve. Python has proven itself over the years to be a serious language, but also as easy as it can be to learn.

- Low floors and high ceilings

# Variables and assignment

It's annoying and boring to use the same values repeatedly. What we want to do is allow for different values in our program. For that, we need *variables*. 

You can think of a variable as a pronoun, referring to a piece of data (a noun). We can assign a value to a variable with the `=` "assignment operator."

A few things to keep in mind:

1. The `=` in Python is **NOT** doing what the `=` in math does. It's very very different.
2. When you assign, Python looks at the right side, and gets the value.
3. It then assigns that value to the variable on the left side.
4. If the variable didn't previously exist, now it does!
5. If the variable *did* previously exist, now it refers to a new value.

```python
x = 10    # here, I'm assigning the value 10 to the variable x

print(x + 5)
```

```python
print(x + 20)
```

```python
print(x + x)
```

In many languages, you need to *declare* your variables. Not so in Python! The first time you assign to a variable is when it is created.

```python
print('abcd' + 'efgh' + 'ijkl')
```

```python
print(10 + 20 + 30)
```

3. Capital letters and lowercase letters are different.
4. Traditionally, we only use lowercase letters (not capitals) in variable names.
5. You shouldn't start or end a variable name with `_`.
6. You cannot start with a digit.



```python
x = 10       # assign the integer (number) 10 to x
x = 'abcd'   # assign the text string 'abcd' to x

print(x)     # the most recent assignment wins!
```

# Conditions

Almost all of programming, when you come down to it, is about making decisions. We can now do that!

We use `if` to make a decision. It looks to its right, and looks for either `True` or `False`. That value (`True` or `False`) usually comes from a comparison operator.

If the `if` condition is `True`, then a block of code runs. But that block of code will *only* run if it's `True`. This is a completely new idea in our programs -- so far, everything has run, from start to finish. But now, only parts of our programs will run.

If the `if` condition is `False`, then a second block, marked by `else`, will run.

One of these blocks will run. Only one of these blocks will run.

```python
# an example

name = input('Enter your name: ')

if name == 'Reuven':
    print('Hello, boss!')
    print('It is great to see you again!')
else:
    print('Hello, ' + name + '.')
```

# What's going on here?

1. We assigned to `name` based on `input`.
2. `if` looks to its right, and sees the result of the `==` comparison between `name` and `'Reuven'`.
3. If it sees a `True` result, then the block starts.
4. Notice that at the end of the `if` line, we have a `:`. This always comes before an indented block.
5. The indented block always comes after a `:`, and can by any number of lines long, and contain any code you want. Lines 6 and 7 are our indented block, and just print values.
6. A block in Python **MUST** be indented, traditionally by 4 spaces. That is how Python knows the block has started, and the lack of indentation means that the block has ended. This is not optional.
7. When you're done with an indented block, just press backspace/delete, and you'll go back to an earlier indentation.
8. An `else` block is optional, but very common. It runs if the `if` condition was `False`.

# What if I have more than two possibilities?

So far, we've thought about using `if`/`else` as mutually exclusive, choosing between two different options.

But what if we have three?

For that, Python provides us with `elif`:

- `elif` must come after an `if` (or an `elif`)
- You can have as many `elif` clauses as you want
- Each has its own comparison, just like `if`
- The first `if`/`elif` comparison to have a `True` result wins, and its block runs
- If none of the `if`/`elif` comparisons are `True`, then the `else` block runs

```python
name  = input('Enter your name: ')

if name == 'Reuven':
    print('Hi again, boss!')

elif name == 'someone else':
    print('That is a very weird name')

else:
    print('Hello, ' + name + '.')
```

```python
x = 50

if x > 20:   # it got here first!  this was the first comparison to return True!
    print('> 20')
elif x > 30:
    print('> 30')
elif x > 40:
    print('> 40')
elif x > 50:
    print('> 50')
else:
    print('Very big')
```

# Is this driving you crazy?

If you're printing things like the above, then you're starting to get frustrated with all of the + signs and the unreadable code.

Fortunately, Python has a solution -- f-strings! These are amazing!

The basic idea of an f-string is:

- Before the opening quote in a string, put an `f`
- Inside of the string, it's normal
- Except wherever you want to put a Python value, put it in `{}`

```python
# let's redo this solution with f-strings

first = input('Enter first word: ')
second = input('Enter second word: ')

if first == second:
    print(f'{first} and {second} are the same!')
elif first < second:
    print(f'{first} comes before {second}')
else:
    print(f'{second} comes before {first}') 
```

```python
# remember how we couldln't put numbers and strings in the same print? 
# f-strings make it possible!

x = 10
y = 20

print(f'{x} + {y}')
```

# Where do things stand?

- We're able to use comparison operators to compare values
- We're able to use `if`/`elif`/`else` to make decisions based on those comparisons, and conditionally execute code depending on what happens there

But... what if our condition is itself pretty complex? What if we want two different things to be `True` in order for something to happen?

For that, we have two special keywords in Python, `and` and `or`:

- `and` combines two different conditions, and returns `True` if both are `True`
- `or` combines two different conditions, and returns `True` if one or both are `True`

```python
x = 10
y = 20

#   True   and   True  --> True
if x == 10 and y == 20:
    print('Both are what you want!')
```

```python
x = 10
y = 300

#    True  and      False  --> False
if x == 10 and y == 20:
    print('Both are what you want!')
```

```python
# what about or? It is a bit more forgiving; it returns True if either condition is True

x = 10
y = 20

#   True   or     True  -> True
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

```python
x = 10
y = 300

#    True   or    False  --> True
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

```python
x = 200
y = 300

#    False  or False --> False
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

A close cousin of `and` and `or` is `not`, which flips the logic from `True` to `False` and back.
