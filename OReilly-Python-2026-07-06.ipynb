- Python was deliberately invented to give everyone the power of a serious programming language without the very difficult learning curve. Python has proven itself over the years to be a serious language, but also as easy as it can be to learn.

A few things to keep in mind:

# Conditions

Almost all of programming, when you come down to it, is about making decisions. We can now do that!

We use `if` to make a decision. It looks to its right, and looks for either `True` or `False`. That value (`True` or `False`) usually comes from a comparison operator.

If the `if` condition is `True`, then a block of code runs. But that block of code will *only* run if it's `True`. This is a completely new idea in our programs -- so far, everything has run, from start to finish. But now, only parts of our programs will run.

If the `if` condition is `False`, then a second block, marked by `else`, will run.

One of these blocks will run. Only one of these blocks will run.

```python
# an example

name = input('Enter your name: ')

if name == 'Reuven':
    print('Hello, boss!')
    print('It is great to see you again!')
else:
    print('Hello, ' + name + '.')
```

# What's going on here?

1. We assigned to `name` based on `input`.
2. `if` looks to its right, and sees the result of the `==` comparison between `name` and `'Reuven'`.
3. If it sees a `True` result, then the block starts.
4. Notice that at the end of the `if` line, we have a `:`. This always comes before an indented block.
5. The indented block always comes after a `:`, and can by any number of lines long, and contain any code you want. Lines 6 and 7 are our indented block, and just print values.
6. A block in Python **MUST** be indented, traditionally by 4 spaces. That is how Python knows the block has started, and the lack of indentation means that the block has ended. This is not optional.
7. When you're done with an indented block, just press backspace/delete, and you'll go back to an earlier indentation.
8. An `else` block is optional, but very common. It runs if the `if` condition was `False`.

# What if I have more than two possibilities?

So far, we've thought about using `if`/`else` as mutually exclusive, choosing between two different options.

But what if we have three?

For that, Python provides us with `elif`:

- `elif` must come after an `if` (or an `elif`)
- You can have as many `elif` clauses as you want
- Each has its own comparison, just like `if`
- The first `if`/`elif` comparison to have a `True` result wins, and its block runs
- If none of the `if`/`elif` comparisons are `True`, then the `else` block runs

```python
name  = input('Enter your name: ')

if name == 'Reuven':
    print('Hi again, boss!')

elif name == 'someone else':
    print('That is a very weird name')

else:
    print('Hello, ' + name + '.')
```

```python
x = 50

if x > 20:   # it got here first!  this was the first comparison to return True!
    print('> 20')
elif x > 30:
    print('> 30')
elif x > 40:
    print('> 40')
elif x > 50:
    print('> 50')
else:
    print('Very big')
```

The basic idea of an f-string is:

- Before the opening quote in a string, put an `f`

# Where do things stand?

- We're able to use comparison operators to compare values
- We're able to use `if`/`elif`/`else` to make decisions based on those comparisons, and conditionally execute code depending on what happens there

But... what if our condition is itself pretty complex? What if we want two different things to be `True` in order for something to happen?

For that, we have two special keywords in Python, `and` and `or`:

- `and` combines two different conditions, and returns `True` if both are `True`
- `or` combines two different conditions, and returns `True` if one or both are `True`

```python
x = 10
y = 20

#   True   and   True  --> True
if x == 10 and y == 20:
    print('Both are what you want!')
```

```python
x = 10
y = 300

#    True  and      False  --> False
if x == 10 and y == 20:
    print('Both are what you want!')
```

```python
# what about or? It is a bit more forgiving; it returns True if either condition is True

x = 10
y = 20

#   True   or     True  -> True
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

```python
x = 10
y = 300

#    True   or    False  --> True
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

```python
x = 200
y = 300

#    False  or False --> False
if x == 10 or y == 20:
    print('At least one is  what you want!')
```

A close cousin of `and` and `or` is `not`, which flips the logic from `True` to `False` and back.

I want you to ask the user to enter a word. You are to print a translation of the word into Pig Latin. (The word will be all lowercase, no spaces, no punctuation.)

The rules of Pig Latin are easy. (Pig Latin is a children's "secret" language.) The idea is that you take an English word, and you look at its first letter:

- If the first letter is a vowel (a, e, i, o, or u) then add "way" to the word
- Otherwise, move the first letter to the end, and add "ay"

Some examples:

- `table` -> `abletay`
- `book` -> `ookbay`
- `papaya` -> `apayapay`
- `elephant` -> `elephantway`
- `octopus` -> `octopusway`

```python
# this works...
word = input('Enter a word: ')

# does it start with a vowel?
if word[0] == 'a' or word[0] == 'e' or word[0] == 'i' or word[0] == 'o' or word[0] == 'u':
    print(word + 'way')
```

```python
# this is more Pythonic

word = input('Enter a word: ')

# does it start with a vowel?
if word[0] in 'aeiou':
    print(word + 'way')
else:
    print(word[1:] + word[0] + 'ay')
```

# Backslashes

Backslashes in strings are used to enter all sorts of special characters:

- `\n` means: Newline, so wherever you see these two characters, when you `print`, it'll go down one line
- `\t` means: Tab, moving to the next multiple-of-8 column on the screen
- `\\` means: One `\` character

```python
s = 'abcdefgh'
print(s)
```

```python
len(s)
```

```python
s = 'abcd\nefgh'
print(s)
```

```python
len(s)
```

```python
print('a\tbcd\tef\tg')
```

```python
# On Windows, if you have a path (folder + filename), that can include \ characters!

path = 'c:\abc\defg\hijkl.txt'

print(path)
```

# What's going on?

1. Modern Python versions warn you if you have a `\` before a character that doesn't do anything meaningful.
2. You can avoid this problem (it tells you) by doubling the backslashes.
3. It also suggests a "raw string," which we'll discuss in a moment.
4. The `\a` goes away, because it's actually a special character, the "alarm bell." 

```python
# option 1: double backslashes

path = 'c:\\abc\\defg\\hijkl.txt'
print(path)
```

```python
# option 2: raw string

# this means: put r before the opening quote
# all this does is double all backslashes

path = r'c:\abc\defg\hijkl.txt'
print(path)
```

# Methods

So far, we've used/seen a bunch of functions:

- `print`
- `input`
- `len`
- `type`

There's nothing wrong with these functions... but Python generally prefers *methods*, which are like functions, except that they are tied to a particular type of value, such as a string. And strings have a *lot* of methods defined on them.

Invoking a method looks a bit different from invoking a function. To call a function, you say:

    FUNC(DATA)

With a method, you say:

    DATA.METHOD()

The fact that METHOD is attached to DATA is a feature, not a bug, because it means you remember what methods are associated with what values. 

When we talk about methods, we must mention them in the context of a particular type. So there isn't a `strip` method, but there is a `str.strip` method.

```python
# let's say that I want to get people's names and greet them

name = input('Enter your name: ')
print(f'Hello, {name}!')
```

Because the string contains lots of leading and trailing whitespace, that's how things are printed.

How can I remove the whitespace? I can invoke a method that does it, `str.strip`.

```python
name.strip()   # invoke str.strip on a string
```

```python
name
```

```python
# string methods return new values. They don't change the original string.
```

```python
# one method of removing whitespace:

name = input('Enter your name: ')  # get input
name = name.strip()                # strip input
print(f'Hello, {name}!')           # print hello with the name
```

```python
# even better: str.strip works on *any* string. 

name = input('Enter your name: ').strip()
print(f'Hello, {name}!')           # print hello with the name
```

# Some other useful string methods

- `str.lower` which returns a new string, all in lowercase, based on the original
- `str.upper` which returns a new string, all in UPPERCASE, based on the original
- `str.isdigit`, which returns `True` if a string is non-empty and contains only digits 0-9.

```python
s = input('Enter your favorite number: ').strip()

if s.isdigit():
    n = int(s)  # no need to worry!
    print(f'Your favorite number * 12345 = {n * 12345}')
else:
    print(f'Not numeric!')
```
