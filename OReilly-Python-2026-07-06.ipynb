# Quick intro to using Jupyter

Everything in Jupyter is done in cells. Every cell has a mode. The most important modes are:

- Markdown, for documentation (i.e., what I'm using now)
- Python, for code



# Writing in Jupyter

When you type into a cell, the cell can be in one of two modes:

- *Edit mode* means that whatever you type is going into the cell. (That's what I'm doing now.) You can enter edit mode by clicking inside of a cell or pressing ENTER.
- *Command mode* means that whatever one character/key you type is used as a command to Jupyter, not displayed on the screen. You can enter command mode by clicking to the left of a cell or pressing ESC.

What commands are available?

- `c` -- copy the current cell
- `x` -- cut the current cell
- `v` -- paste the most recently cut/copied cell
- `a` -- new cell *above* the current one
- `b` -- new cell *belowW* the current one
- `y` -- make the current cell Python
- `m` -- make the current cell Markdown documentation

When you're done editing a cell, press SHIFT+ENTER and it'll be run (if Python) or formatted (if Markdown)

# What is a programming language?

We're here to learn Python. Python is a programming language. But what does that even mean?

When computers were first invented, they could solve one problem at a time. If you wanted to change the problem that a computer was solving, you needed to rebuild the computer, or get a new one. 

Pretty soon, people realized that it would make more sense to create a general-purpose, programmable computer that could solve many different types of problems. How did we give it instructions? We would use 1s and 0s, the so-called "binary code."

Programming in binary was very difficult and tedious (and error prone!). Pretty soon, people created programming languages. The idea was that you would write in a language, and then you would use a program to translate your program into 1s and 0s. You can think of this as like getting a book translated.

The programming languages that people created had different advantages and disadvantages. Some ran really fast. Some were easy to understand. Some were great for math. Some were great for graphics. etc.

There are, today, many thousands of programming languages. Some are:

- C and C++: These run super fast, in part because they are very close to the 1s and 0s, making it hard to write with these languages, but they're super efficient.
- Java and C#: Easier to work with than C and C++, but not quite as fast. Also, automatic management of memory, which is a big deal, since in C, you have to keep track of what memory you're using.
- Python was deliberately invented to give everyone the power of a serious programming language without the very difficult learning curve. Python has proven itself over the years to be a serious language, but also as easy as it can be to learn.

- Low floors and high ceilings
