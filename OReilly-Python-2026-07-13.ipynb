# DRY -- don't repeat yourself

One of the most important rules in programming is indeed "DRY -- don't repeat yourself." This means that if you have several lines of code in a row that are the same or almost the same, you almost certainly don't want to be writing the code in this way. There must be a better way to do it.

Often, the best way is to use a *loop* -- the loop will allow us to repeat an action, or a set of actions, but we only have to write that action a single time. The program (computer) will repeat things for us.

Programming languages give us many different kinds of loops. In Python, we have two:

- `for` loops, more common and very useful in going over strings and similar sequences
- `while` loops, which are basically `if` statements that repeat themselves so long as the condition is `True`

```python
s = 'abcd'

print('Before')
for one_character in s:
    print(one_character)
print('After')    
```

# Syntax of a `for` loop

- It starts with the keyword `for`
- Then we have our *loop variable*. This is where we tell the loop what variable should be assigned a new value with each iteration of the loop.
- Then we have the keyword `in`
- Then we have the value over which we want to iterate. In this case, that would `s`, the string. (Or a variable referring to a string.)
- At the end of the line, we have a `:` -- because then we have an indented loop body, a Python block.
- The loop body can be as long or as short as you want, and it can contain any Python code that you want. Typically, you want to keep it short and clear.
- The loop wil execute once for each item that the value (`s`, in this case) provides. Once we're done getting values from `s`, the loop exits and we continue executing our Python program after the loop body.

# The `for` conversation -- what's really going on?

- `for` turns to `s`, the value at the end of the line, and asks: Are you iterable? Do you know how to even behave inside of a `for` loop?
    - If not, then the loop exits with an error (TypeError), saying that this kind of value isn't iterable
- If the value is iterable, then `for` asks it for its next value.
    - If there is no next value, then the loop exits
- If there is a next value, then it is assigned to the loop variable -- `one_character` in this example.
- You can call your loop variable whatever you want. It has ZERO effect on what values you get in each iteration, or how many there are.
- When `for` asks for the next value from `s`, it has no idea how many values we still have to get. It also doesn't know what kind of value it's going to get.
- When we reach the end of the loop body, we return to step 2, where we ask for the next value

Notice what we don't have:
- We don't know what types will be returned
- We don't know how many values will be returned
- We don't get the index of the value that we got
- We're not setting an index, either.

Also: The loop variable's final assignment before the loop ends remains with the program even after we have exited the loop.

# Exercise: Vowels, digits, and others

We're going to ask the user to enter a text string. We're then going to go through that string, one character at a time, and check: 

- Is it a vowel (a, e, i, o, or u)? If so, then add 1 to the `vowels` counter
- Is it a digit (0-9)? If so, then add 1 to the `digits` counter
- In all other cases, add 1 to the `others` counter

1. Define three variables, `vowels`, `digits`, and `others`, and set them all to be 0.
2. Ask the user, with `input`, to enter some text.
3. Use a `for` loop to go through the text, one character at a time
    - Increment the appropriate variable/counter
4. Print each of the counters.

```python
vowels = 0
digits = 0
others = 0

text = input('Enter text: ').strip()

for one_character in text.lower():
    if one_character in 'aeiou':     # if it's a vowel?
        vowels += 1                  # this means take the current value of vowels, add 1, and assign back to vowels
    elif one_character.isdecimal():  # if it's a digit?
        digits += 1
    else:
        others += 1

print(f'{vowels=}')           # cool trick; in an f-string, in {}, if you say f'{x=}', you'll get x=10
print(f'{digits=}')   
print(f'{others=}')   
```

# Where's the index?

In a language like C, you don't iterate over the characters of a string. Rather, you iterate over the indexes of the string's characters, and use the index to retrieve the current character. You (writing the `for` loop) are responsible for starting the index at the right point (usually 0), ending the loop before the index goes out of bounds, and retrieving values from the string to work.

Short version: In C, you have to keep track of the index, and use it to retrieve characters.

in Python, though, you just get the characters, without having to work hard. Who needs the index?

Most of the time, the answer is: No one.

However, sometimes we do want it.

How can I get the index, say, to print it next to each character?

```python
# option 1: Do it manually

text = 'abcde'
index = 0   # we are in charge of the index

for one_character in text:
    print(f'{index}: {one_character}')   # print the current index/one_character
    index += 1                           # add 1 to the index
```

```python
# option 2: Do it automatically, with enumerate
# enumerate is a function that "wraps" around an existing iterable
# it ensures that we get TWO variables returned for each iteration -- the index and the value

text = 'abcde'

for index, one_character in enumerate(text):
    print(f'{index}: {one_character}')   # print the current index/one_character
```

# Controlling our loops

Sometimes, we are in a `for` loop, and we realize that we've accomplished what we wanted.

This means, sometimes,  that we want to tell the loop to exit prematurely -- wherever you are, whatever you're doing, jump to the end of the loop.

That is the `break` command. It tells Python that you don't want to run the current loop any more. The program will continue with the next command after the loop body finishes executing.

Why would I want `break`? Mostly beacuse you've already accomplished the goal, and going through the loop again is a waste of time and money.

```python
s = 'abcde'

print('Start')
for one_character in s:

    if one_character == 'c':
        break

    print(one_character)
print('Done')
```

# The `continue` command

`continue` means: I don't have anything more to do with this iteration. But maybe the next iteration will be better. So finish this one, and go onto the next one immediately.

Technically speaking: Go, continue onto the next iteration.



```python
s = 'abcde'

print('Start')
for one_character in s:

    if one_character == 'c':
        continue

    print(one_character)
print('Done')
```

# Summarizing `break` and `continue`

Normally, both are run in a loop body, inside of an `if` statement.

- If you have accomplished your goals, `break` from the loop
- If you don't need the current iteration, but want to give others a chance, use `continue` to force the `for` loop onto the next iteration (assuming there is one).

# Exercise: Sum digits

1. Ask the user to enter a string with digits.
2. Set `total` to 0.
3. Go through each character that was entered:
    - If it's a `.`, then then stop the loop entirely, and print `total`
    - If it's a non-digit (other than `.`), then scold the user and go on.
    - If it's a digit, then add to `total`
4. Print `total`.

My favorite example of where to use `continue` is when you're reading through a file, one line at a time, and the line starts with `#`. You can use `continue`, because the whole point of comments is to ignore them. So if a Python program is reading such a file, and encounters `#` on each line, that's fine -- each needs to be dealt with.

```python
total = 0

s = '23a4.b5'

for one_character in s:
    if one_character == '.':   # . charcter? end the loop!
        print(f'\tGot a . -- ending')
        break

    if not one_character.isdecimal():  # non-digit? end this iteration
        print(f'\tHey! {one_character} is not a number!')
        continue

    # if I got here, I know that one_character is a digit
    # I know that I can turn into an integer
    total += int(one_character)   # add the digit to total
    print(f'\t{total=}')

print(total)    
```

# `.isdecimal` vs. `.isdigit`

Both of these (as well as `isnumeric`) are methods you can run on strings. They all do mostly the same thing. 

- `isnumeric` will answer `True` not only for digits 0-9, but also numbers in other character sets, such as Arabic and Chinese
- `isdigit` will answer `True` not only for regular numbers, but also for superscripts and subscripts (e.g., 2 to the 3rd power or O2 in chemistry)
- `isdecimal` just returns `True` if you have "normal" numbers.

# `while`

So far, we've looked at `for` loops. They run over the course of a series of values -- all of the values in a string, or all of the values in a range, etc.

You know how many iterations will be needed for a `for` loop. (The maximum number of iterations, that is...)

But what if you don't know how long it'll take to accomplish your goal? What if it might take 2 iterations, but it might also take 200? 

In such a case, you need a way to loop that isn't dependent on a number, but is on the results. So long as we haven't yet achieved the results we want, we'll keep iterating.

That is what a `while` loop does. It is like an `if` that keeps coming back to the condition -- so long as the condition is `True`, the `while` loop's body executes. If/when it's `False`, the loop exits.

You can express many things in either a `while` or a `for` loop. But it's more natural (usually) to choose.

```python
# while loop example

x = 5

while x > 0:   # so long as x > 0...
    print(x)   # print x
    x -= 1     # decrement x by 1
print('Done') 
```

# Exercise: Sum to 100

1. Set `total` to 0
2. Ask the user to enter a number.
     - If it's not a valid number, then scold the user.
     - If it's the empty string, then stop the program
     - Convert the number to an `int` and add to `total`
3. When `total` is >= 100, stop, and print `total`

```python
total = 0

while total < 100:
    number = input('Enter a number: ').strip()

    if number == '':   # break if we got the empty string
        break

    if not number.isdecimal():  # not a number?
        print(f'{number} is not numeric; try again')
        continue

    total += int(number)
    print(f'{total=}')

print(f'In the end, total = {total}')    
    
```

# Lists

Lists are another data type in Python -- one of the core, and most important ones. A list is a "collections," meaning that it's not just important on its own. It's important because it can contain other values.

How many values? As many as you want. (Or as few as you want.)

What kinds of values? Whatever you want, in any combination -- although it's traditional to only have one type of value in a list. So it's normal to have a list of strings, a list of integers, a list of lists, etc.

Where are lists, used? All over the place -- a list of database records. A list of filenames. A list of functions. A list of variables. 

Lists are often called Python's "default container class," meaning that if have to keep track of a number of values, you'll often put them into a list.

Lists aren't arrays, for at least two reasons:

1. Lists can contain values of different types, and arrays (in theory) can only contain values of one type
2. Lists can expand and contract, changing the number of values they contain. Arrays (in theory) cannot change their size once created.

There are plenty of languages that have arrays, and which violate one or both of these. However, Python talks about lists, and if you mention arrays, it'll likely be in the context of NumPy arrays, used for analyzing data.
