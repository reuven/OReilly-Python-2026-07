# Dictionaries ("dicts")

We've seen in the last two weeks a number of different Python data structures.

- Numbers (both `int` and `float`)
- Text (`str`)
- Container sequences (`list` and `tuple`)

In the cases of `str`, `list`, and `tuple`, we have an index that starts at 0 and goes up to `len(s) - 1`. That's not bad, but it's frustrating sometimes to have the index be completely unconnected from the data itself.

One way to think of dicts is that they are like lists, except that the index is determined by *you*. No longer is the index 0, 1, 2, 3, etc. Rather, you can use just about any "key" that you want for storing and retrieving your data. 

This makes them very convenient. But they also happen to be very efficient -- very fast search and very fast retrieval.

# Dicts aren't unique to Python

Other languages have used them for many years, too. They normally have other names:

- Hash tables
- Hash maps
- Maps
- Hashes
- Key-value stores
- Name-value stores
- Associative arrays

No matter what you call them, dicts are stored as pairs -- a key-value pair. Unlike a list, where you have a value at a particular index, and the index is kind of incidental, in a dict, the key and the value are married to one another. You use the key to retrieve (or set) the value. 

# To create a dict

1. We use `{}`
2. The key and value are separated from one another with `:`
3. Each key-value pair is separated from other key-value pairs with `,`

```python
d = {'a':10, 'b':20, 'c':30}   # 3 key-value pairs

d = {}   # empty dict, with zero key-value pairs
```

# Rules for creating dicts

1. A dict can contain any number of key-value pairs. The minimum is zero (the empty dict), and the maximum is determined by your computer's memory.
2. Any immutable value in Python (e.g., numbers and strings and tuples) can be a key, so long as it doesn't contain a mutable value. Normally, we use numbers and strings as dict keys.
3. In any given dict, the keys must be unique.
4. Every key has a value, and every value has a key.
5. There are *NO RESTRICTIONS WHATSOEVER* on values. They can be mutable. They can repeat themselves. etc.

```python
d = {'a':10, 'b':20, 'c':30}

len(d)  # length is 3, because we're counting pairs
```

```python
# retrieve the value associated with the key 'a'
# use []

d['a']   
```

```python
d['c']
```

```python
d['x']
```

```python
# remember that keys must match *exactly*!

d['C']   # capital c
```

```python
# what about whitespace?

d['b ']
```

# Can I use a variable as a key?

If the variable refers to an immutable value -- a string or number -- then yes, you can use the variable, because it'll be replaced by the value when Python sets or retrieves the values.

```python
d
```

```python
k = 'b'

d[k]  # Python first evaluates k into 'b', then asks for d['b'], which happens to be 20
```

```python
# I can check if a key is in a dict with the "in" operator
# it *only* checks the keys, not the values! 

'a' in d   # really: is 'a' a key in d?
```

```python
'x' in d   # is 'x' a key in d?
```

```python
20 in d  # is 20 a key in d?
```

Remember that when working with dicts, the keys are primary. We use the keys to do everything, and the values are kind of along for the ride.

# Where are dicts useful?

Answer: Everywhere. 

- A dict whose keys are the names of the months of the year, and the values are the numbers of those months.
- A dict whose keys are the numbers of the months of the year, and the values are the names of those months.
