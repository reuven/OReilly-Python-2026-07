# Dictionaries ("dicts")

We've seen in the last two weeks a number of different Python data structures.

- Numbers (both `int` and `float`)
- Text (`str`)
- Container sequences (`list` and `tuple`)

In the cases of `str`, `list`, and `tuple`, we have an index that starts at 0 and goes up to `len(s) - 1`. That's not bad, but it's frustrating sometimes to have the index be completely unconnected from the data itself.

One way to think of dicts is that they are like lists, except that the index is determined by *you*. No longer is the index 0, 1, 2, 3, etc. Rather, you can use just about any "key" that you want for storing and retrieving your data. 

This makes them very convenient. But they also happen to be very efficient -- very fast search and very fast retrieval.

# Dicts aren't unique to Python

Other languages have used them for many years, too. They normally have other names:

- Hash tables
- Hash maps
- Maps
- Hashes
- Key-value stores
- Name-value stores
- Associative arrays

No matter what you call them, dicts are stored as pairs -- a key-value pair. Unlike a list, where you have a value at a particular index, and the index is kind of incidental, in a dict, the key and the value are married to one another. You use the key to retrieve (or set) the value. 

# To create a dict

1. We use `{}`
2. The key and value are separated from one another with `:`
3. Each key-value pair is separated from other key-value pairs with `,`

````python
d = {'a':10, 'b':20, 'c':30}   # 3 key-value pairs

d = {}   # empty dict, with zero key-value pairs
````

# Rules for creating dicts

1. A dict can contain any number of key-value pairs. The minimum is zero (the empty dict), and the maximum is determined by your computer's memory.
2. Any immutable value in Python (e.g., numbers and strings and tuples) can be a key, so long as it doesn't contain a mutable value. Normally, we use numbers and strings as dict keys.
3. In any given dict, the keys must be unique.
4. Every key has a value, and every value has a key.
5. There are *NO RESTRICTIONS WHATSOEVER* on values. They can be mutable. They can repeat themselves. etc.

````python
d = {'a':10, 'b':20, 'c':30}

len(d)  # length is 3, because we're counting pairs
````

````python
menu = {'sandwich':10, 'tea':5, 'apple':1, 'cake':3}
total = 0

order = input('Order: ').strip()

if order in menu:             # if the user's input is a key in our "menu" dict...
    price = menu[order]       # get the price (i.e., the value associated with the key for this item)
    total += price            # add the price to the total
    print(f'{order} costs {price}; total is now {total}')

else:                         # user's string/input is not a key in our menu dict, so scold the user
    print(f'We are out of {order} today -- try something else!')

print(f'total = {total}')    
````

````python
# we need to add a loop -- let the user order more than one thing!

menu = {'sandwich':10, 'tea':5, 'apple':1, 'cake':3}
total = 0

while True:  # infinite loop!
    order = input('Order: ').strip()

    if order == '':              # no order at all?
        break                    # break out of the while loop
    
    if order in menu:             # if the user's input is a key in our "menu" dict...
        price = menu[order]       # get the price (i.e., the value associated with the key for this item)
        total += price            # add the price to the total
        print(f'{order} costs {price}; total is now {total}')
    
    else:                         # user's string/input is not a key in our menu dict, so scold the user
        print(f'We are out of {order} today -- try something else!')

print(f'total = {total}')    
````

# Exercise: Vowels, digits, and others (dict edition)

1. Define a dict, `counts`, with three keys -- `vowels`, `digits`, and `others`. All three values for these keys should be 0.
2. Ask the user to enter a string.
3. Go through that string, one character at a time.
    - If it's a vowel, add 1 to the value for `vowels`
    - If it's a digit, then add 1 to the value of `digits`
    - Otherwise, add it to `others`
4. Print `counts` in the end.

````python
counts = {'vowels': 0,
          'digits': 0,
          'others': 0}

text = input('Enter text: ').strip()

for one_character in text:
    if one_character in 'aeiou':    # if the current character is a vowel...
        counts['vowels'] += 1       # ... add 1 to the existing value of counts['vowels']
    elif one_character.isdecimal(): # if the current character is a decimal...
        counts['digits'] += 1       # ... add 1 to the existing value of counts['digits']
    else:
        counts['others'] += 1       # anything else? just add 1 to 'others'

print(counts)        
````

# Three paradigms of dict usage

There are *many* ways to use dicts in Python. Three of the most common are as follows:

1. Define a dict, and *never* change it. Instead, treat it as an immutable, read-only database in memory. (Restaurant)
2. Define a dict with keys and initial values. The keys won't change; we won't add or remove any keys once the program starts. But the values will be changed and updated to reflect accumulation of data we've seen in the life of the program. (Vowels, digts, and others)
3. Define an empty dict, aka "counting the unknown." We can (and will) add new keys over time to reflect new data. We'll also update the values as needed. This is great when we don't know what keys or values we'll have, but we want to measure a wide variety of values.

````python
# example: count characters in text

counts = {'a':0, 'b':0, 'c':0, 'd':0, 'e':0}

text = input('Enter text: ').strip()

for one_character in text:
    counts[one_character] += 1

print(counts)    
````

If I want the above code to work, I need to define `counts` with *every* possible character as a key. All letters, numbers, symbols, emojis, etc.

It's just not worth doing that. Instead, we can define `counts` as an empty dict, adding new key-value pairs as we go through it.

````python
# version 2

counts = {}  # empty dict -- who knows what we'll be counting?

text = input('Enter text: ').strip()

for one_character in text:
    if one_character in counts:     # have we already seen one_character before?
        counts[one_character] += 1  # ... if so, then yeah, add 1 to the count
    else:
        counts[one_character] = 1   # ... if not, then add a new key-value pair with one_character and a starting count of 1

print(counts)    
````

# Exercise: Rainfall

The goal of this exercise is to keep track of the rainfall in different cities over time. We don't know what cities we'll be measuring. We'll rely on the user to enter the name of a city, and then to tell us how much rain (in mm) fell there most recently. The keys of the dict will be strings, the city names. The values will be integers.

1. Define an empty dict, `rainfall`.
2. Ask the user repeatedly to enter the name of a city.
    - If the user enters an empty city name, break from the loop
3. If you got a city name, then ask the user to enter mm rain that fell
    - Let's assume that you get digits
4. If you have data for this city from before, add this new amount to the existing amount
5. If you haven't seen this city before, then add it (and this amount) to the `rainfall` dict.
6. When the user finishes entering data, print the entire `rainfall` dict.

````python
rainfall = {}

while True:
    city_name = input('City: ').strip()

    if city_name == '':    # no city name? Exit the loop!
        break
    
    mm_rain = input('Rain: ').strip()
    mm_rain = int(mm_rain)
    
    if city_name in rainfall:             # if we have seen this city before...
        rainfall[city_name] += mm_rain    # ... add mm_rain to the previously recorded rainfall
    else:    
        rainfall[city_name] = mm_rain     #if this is the first time we're seeing a city, add a new key-value pair to the dict

print(rainfall)
````

# How about a variation on this?

Instead of simply totaling the `mm_rain` values, how about the dict values contain lists of integers? Then we can provide not only the total rainfall but also the mean rainfall per city. Instead of writing all of the code ourselves, we can even use Python's builtin `sum` function, which takes a list as an argument, and returns the sum.

````python
rainfall = {}

while True:
    city_name = input('City: ').strip()

    if city_name == '':    # no city name? Exit the loop!
        break
    
    mm_rain = input('Rain: ').strip()
    mm_rain = int(mm_rain)
    
    if city_name in rainfall:                  # if we have seen this city before...
        rainfall[city_name].append(mm_rain)    # ... append mm_rain to the previously recorded rainfall (list)
    else:    
        rainfall[city_name] = [mm_rain]        # if this is the first time we're seeing a city, add a new key-value pair, with a str + list

print(rainfall)
````

# Iteration and dicts

We've seen that we can iterate (with `for`) over a number of different data types:

1. If we iterate over a string, we get each character
2. Iterating over a `range` gives us integers
3. Itrerating over a list or tuple gives us elements

What happens if we iterate a dict?

````python
d = {'a':10, 'b':20, 'c':30}

for one_thing in d:   # iterating over a dict gives us the keys, one at a time
    print(one_thing)
````

````python
d = {'a':10, 'b':20, 'c':30}

for one_key in d:   
    print(f'{one_key}: {d[one_key]}')
````

There is a dict method, `dict.keys`, which returns the keys when you invoke it. There is almost no reason to invoke `dict.keys`. Most of the time, you want to search on the dict itself (with `in`) and you want to iterate over the keys (with just the dict).

However. It's kind of weird and annoying that I get the key, and then need to use the dict to retrieve the value.

It would be nice to get the key and value together in each iteration.

That's the `dict.items` method - which returns a 2-element `(key, value)` tuple with each iteration.

````python
for key, value in d.items():   # each iteration will give us a key-value tuple
    print(f'{key}: {value}')
````

# How do dicts work?

Yes, dictionaries are closer to the real world than lists and tuples, because they use "real" keys rather than numeric indexes that start with 0. 

It turns out that dicts are great not just because keys can be any numbers or text we want. They're also much faster at finding and retrieving information than lists are. If we look for a value in a list, then we don't know if the value will be found right away, close to the end, or never. The more values you have in a list, the longer it'll take, on average, to find one.

By contrast, dicts are almost immediate in their ability to find and retrieve values based on the key. It doesn't matter how many key-value pairs you have in your dict; Python, given a key, can tell you if it's in the dict and retrieve its value right away.

The way this is possible is with a special function known as a "hash function." 

How does this work? How does this help?

If I store a new value (using `list.append`) in a list, Python stores that value at the next empty location. That's why we need to search through a list for its values to see if something is there.

By contrast, if I store a key-value pair in a dict, the location of that storage is based on the key. Python runs `hash(key)`, which gives a number back. The number isn't random, but it looks kind of random, and you cannot predict its value based on the key. 

We ask Python to store a key-value pair; it does so at the memory location returned by `hash(key)`. If we ask `key in d`, Python runs `hash(key)`, jumps to that location in memory, and can tell us if the key is there.

If it's not there, we'll know right away.

If the key is there, then we can get the value back without much more effort.

When you assign to a variable in Python, the variable name is turned into a string, and is used as a key in a dict of variables! All variables in Python are really dict keys. And the variable values are the dict values.

Why, then, can we not use mutable values (e.g., lists) as dict keys? What's wrong with that?

The problem is that if you use `hash(key)` to figure out where the key-value pair should be stored, and then `key` changes, suddenly `hash(key)` is likely different, too. To avoid that sort of "missing key-value pair" situation, Python just refuses to allow us to use mutable data (or things containing mutable data) as dict keys.

# Files

We use files all of the time in our modern, computer-centric world:

- Word files
- Powerpoint files
- PDF files
- JPEG and PNG files

What is a file? Why do we have them? What are they useful for?

Technically, a file is a bunch of bytes (kind of like letters) stored on disk. The idea is that a program works with data structures in memory. If you restart the computer, the memory is wiped. If you want to move those data structures from one computer to another, you'll need a way to do it.

A file is a way to turn data structures in memory into a permanent form. That permanent form survives restarts, and also allows you to transmit it to another computer. 

What is a Word document? When you save a Word document, you're translating the data structures in memory to something portable and permanent. 

When you want to read your Word document into Word, it turns the portable, permanent version into something in memory that you can use.

When we want to read a file into Python, we'll need to know how to translate from a binary format into a bunch of data structures in memory. But... we're going to simplify things a lot, and only think about text files. 

Text files are especially popular among system and network administrators, where they can be logfiles or config files.
