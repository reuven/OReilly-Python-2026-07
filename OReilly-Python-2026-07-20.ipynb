# Dictionaries ("dicts")

We've seen in the last two weeks a number of different Python data structures.

- Numbers (both `int` and `float`)
- Text (`str`)
- Container sequences (`list` and `tuple`)

In the cases of `str`, `list`, and `tuple`, we have an index that starts at 0 and goes up to `len(s) - 1`. That's not bad, but it's frustrating sometimes to have the index be completely unconnected from the data itself.

One way to think of dicts is that they are like lists, except that the index is determined by *you*. No longer is the index 0, 1, 2, 3, etc. Rather, you can use just about any "key" that you want for storing and retrieving your data. 

This makes them very convenient. But they also happen to be very efficient -- very fast search and very fast retrieval.

# Dicts aren't unique to Python

Other languages have used them for many years, too. They normally have other names:

- Hash tables
- Hash maps
- Maps
- Hashes
- Key-value stores
- Name-value stores
- Associative arrays

No matter what you call them, dicts are stored as pairs -- a key-value pair. Unlike a list, where you have a value at a particular index, and the index is kind of incidental, in a dict, the key and the value are married to one another. You use the key to retrieve (or set) the value. 

# To create a dict

1. We use `{}`
2. The key and value are separated from one another with `:`
3. Each key-value pair is separated from other key-value pairs with `,`

````python
d = {'a':10, 'b':20, 'c':30}   # 3 key-value pairs

d = {}   # empty dict, with zero key-value pairs
````

# Rules for creating dicts

1. A dict can contain any number of key-value pairs. The minimum is zero (the empty dict), and the maximum is determined by your computer's memory.
2. Any immutable value in Python (e.g., numbers and strings and tuples) can be a key, so long as it doesn't contain a mutable value. Normally, we use numbers and strings as dict keys.
3. In any given dict, the keys must be unique.
4. Every key has a value, and every value has a key.
5. There are *NO RESTRICTIONS WHATSOEVER* on values. They can be mutable. They can repeat themselves. etc.

````python
d = {'a':10, 'b':20, 'c':30}

len(d)  # length is 3, because we're counting pairs
````

````python
# retrieve the value associated with the key 'a'
# use []

d['a']   
````

````python
d['c']
````

````python
d['x']
````

````python
# remember that keys must match *exactly*!

d['C']   # capital c
````

````python
# what about whitespace?

d['b ']
````

# Can I use a variable as a key?

If the variable refers to an immutable value -- a string or number -- then yes, you can use the variable, because it'll be replaced by the value when Python sets or retrieves the values.

````python
d
````

````python
k = 'b'

d[k]  # Python first evaluates k into 'b', then asks for d['b'], which happens to be 20
````

````python
# I can check if a key is in a dict with the "in" operator
# it *only* checks the keys, not the values! 

'a' in d   # really: is 'a' a key in d?
````

````python
'x' in d   # is 'x' a key in d?
````

````python
20 in d  # is 20 a key in d?
````

Remember that when working with dicts, the keys are primary. We use the keys to do everything, and the values are kind of along for the ride.

# Exercise: Restaurant

1. Define a dict, `menu`, which will contain key-value pairs describing a restaurant menu. The keys will be the menu items, and the values will be the prices.
2. Define `total` to be 0.
3. Ask the user repeatedly to order an item from the menu.
    - If the user enters an empty string, that signals they are done ordering. Print their total.
    - If the user enters a string that is a menu key, then print the item, the price, and the new total.
    - If the user enters a string that is not a menu key, then scold the user, and ask again.
4. Print `total`.

```
Order: sandwich
sandwich costs 10, total is 10
Order: tea
tea costs 5, total is 15
Order: elephant
Sorry, we're fresh out of elephant today!
Order: [ENTER]

total is 15
```

````python
menu = {'sandwich':10, 'tea':5, 'apple':1, 'cake':3}
total = 0

order = input('Order: ').strip()

if order in menu:             # if the user's input is a key in our "menu" dict...
    price = menu[order]       # get the price (i.e., the value associated with the key for this item)
    total += price            # add the price to the total
    print(f'{order} costs {price}; total is now {total}')

else:                         # user's string/input is not a key in our menu dict, so scold the user
    print(f'We are out of {order} today -- try something else!')

print(f'total = {total}')    
````

````python
type(menu)
````

````python
# we need to add a loop -- let the user order more than one thing!

menu = {'sandwich':10, 'tea':5, 'apple':1, 'cake':3}
total = 0

while True:  # infinite loop!
    order = input('Order: ').strip()

    if order == '':              # no order at all?
        break                    # break out of the while loop
    
    if order in menu:             # if the user's input is a key in our "menu" dict...
        price = menu[order]       # get the price (i.e., the value associated with the key for this item)
        total += price            # add the price to the total
        print(f'{order} costs {price}; total is now {total}')
    
    else:                         # user's string/input is not a key in our menu dict, so scold the user
        print(f'We are out of {order} today -- try something else!')

print(f'total = {total}')    
````

# Are dicts mutable?

We've seen that strings are immutable -- they cannot be changed once created -- but lists are mutable. We've seen that mutating a value involves three possible methods:

- You can update an existing element
- You can add a new element
- You can remove an existing element

How does this work with dicts?

````python
d = {'a':10, 'b':20, 'c':30}

# how can I replace c's 30 value with my 12345 value?
# answer: just assign via the key!

d['c'] = 12345
````

````python
d
````

````python
# what about adding a new key-value pair?
# maybe we need a special method, like append.

# it's super simple, much easier than with lists, to add a new key-value pair
# technique: Just assign.

# it's the same syntax to update an existing value as it is to assign a new key-value pair

d['x'] = 98765
d
````

I can just add a new key-value pair to any Python program, anything at all?

Yes. 

Whenever you want to add a new key-value pair to a dict, just assign to it. 

If you want to replace an existing value, just assign to it.



````python
# To remove a key-value pair, use the dict.pop method, passing it the key
# it will (a) remove the pair from the dict and (b) return the value of the removed pair
````

````python
d
````

````python
d.pop('x')  # remove the key-value pair of 'x'/98765
````

````python
d.pop('x')
````

# Exercise: Vowels, digits, and others (dict edition)

1. Define a dict, `counts`, with three keys -- `vowels`, `digits`, and `others`. All three values for these keys should be 0.
2. Ask the user to enter a string.
3. Go through that string, one character at a time.
    - If it's a vowel, add 1 to the value for `vowels`
    - If it's a digit, then add 1 to the value of `digits`
    - Otherwise, add it to `others`
4. Print `counts` in the end.

````python
counts = {'vowels': 0,
          'digits': 0,
          'others': 0}

text = input('Enter text: ').strip()

for one_character in text:
    if one_character in 'aeiou':    # if the current character is a vowel...
        counts['vowels'] += 1       # ... add 1 to the existing value of counts['vowels']
    elif one_character.isdecimal(): # if the current character is a decimal...
        counts['digits'] += 1       # ... add 1 to the existing value of counts['digits']
    else:
        counts['others'] += 1       # anything else? just add 1 to 'others'

print(counts)        
````
